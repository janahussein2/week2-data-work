```python
def extract_listing_data(soup_obj):
    """Extract all data from a listing content div"""
    data = {}
    
    # 1. TITLE
    title_elem = soup_obj.select_one('._titleRow__1AWv1 h4')
    data['title'] = title_elem.get_text(strip=True) if title_elem else ''
    
    # 2. PRICE
    price_elem = soup_obj.select_one('._price__X51mi span')
    if price_elem:
        # Remove inner spans if they exist
        for span in price_elem.find_all('span'):
            span.decompose()
        data['price'] = price_elem.get_text(strip=True)
    else:
        data['price'] = ''
    
    # 3. SPECIFICATIONS
    specs = soup_obj.select('._specs__nbsgm ._spec__SIJiK')
    
    # Clean each spec (remove span and img tags)
    clean_specs = []
    for spec in specs[:3]:  # Get only first 3
        # Remove span and img tags to get clean text
        for tag in spec.find_all(['span', 'img']):
            tag.decompose()
        clean_text = spec.get_text(strip=True)
        clean_specs.append(clean_text)
    
    # Fill with empty strings if less than 3 specs
    while len(clean_specs) < 3:
        clean_specs.append('')
    
    data['area'] = clean_specs[0]
    data['bedrooms'] = clean_specs[1]
    data['bathrooms'] = clean_specs[2]
    
    # 4. DESCRIPTION
    desc_elem = soup_obj.select_one('._description__zVaD6 p')
    if desc_elem:
        # Clean description - remove extra whitespace
        description = ' '.join(desc_elem.get_text().split())
        data['description'] = description
    else:
        data['description'] = ''
    
    # 5. LOCATION
    loc_elem = soup_obj.select_one('._footer__CnldH p')
    if loc_elem:
        # Combine all text in location
        location = loc_elem.get_text(strip=True)
        data['location'] = location
    else:
        data['location'] = ''
    
    return data

# Extract data
listing_data = extract_listing_data(soup)

# Print beautifully
print("🏠 EXTRACTED LISTING DATA")
print("=" * 80)

for key, value in listing_data.items():
    if key == 'description' and len(value) > 100:
        print(f"\n📝 {key.upper()}:")
        print(f"{value[:200]}...")
    else:
        print(f"\n{key.upper()}: {value}")

print("\n" + "=" * 80)
```

```python
import httpx
from bs4 import BeautifulSoup
import json

def scrape_all_listings():
    """Scrape ALL listings from the page"""
    
    # Fetch the page
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
    }
    
    url = "https://sa.aqar.fm/"
    response = httpx.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find ALL content divs (all listings)
    all_content_divs = soup.select('._content__W4gas')
    
    print(f"Found {len(all_content_divs)} listing content sections")
    
    # Create list of dictionaries
    list_of_dicts = []
    
    for i, content_div in enumerate(all_content_divs, 1):
        try:
            data = extract_listing_data(content_div)
            data['listing_number'] = i  # Add sequence number
            list_of_dicts.append(data)
            print(f"✓ Added listing #{i}")
        except Exception as e:
            print(f"✗ Error with listing #{i}: {e}")
            continue
    
    return list_of_dicts

# Get all listings
all_listings = scrape_all_listings()

print(f"\n✅ Created list with {len(all_listings)} dictionaries")
```
